```python
#a)
def f(c,x):
    return 1 - exp(-c*x) 

c = 2
x=1
tol = 1e-6
erro = 1000

while (erro>tol):
    xold = x
    x = f(c,x)
    erro = abs(x-xold)

print(f'A solução é x = {x:.2f}')
```

```python
#b)
c = np.arange(0,3.01,0.01)
raizes = []

for i in range(0,len(c)):
    x = 1
    erro = 1000
    while (erro>tol):
        xold = x
        x = f(c[i], x)
        erro = abs(x-xold)
    raizes.append(x)

print(f"A equação possui {len(raizes)} raizes")

plt.plot(c,raizes, color = "c")
plt.xlabel("c")
plt.ylabel("x")
```

### Questão 2

 a) Partimos das equações:
$$
-x+ay+x²y = 0 \quad (1) \qquad , \qquad b- ay - x²y = 0 \quad (2)
$$
Da equação (2) tiramos que $x²y = b - ay$. Então, substituindo isto em (1), temos:
$$
- x + ay + b - ay = 0 \qquad \rightarrow \qquad x = b
$$
Agora que achamos x, substituímos na equação 2 e temos:
$$
b²y - b + ay = 0 \qquad \rightarrow \qquad y (a + b²) = b \qquad \rightarrow \qquad y = \frac{b}{a+b²} \quad (a)
$$



b)Do item anterior, se substituirmos x no lugar do b em (a), temos:
$$
y = \frac{b}{a+b²}  \qquad \rightarrow \qquad \frac{x}{a+x²} = y \qquad \rightarrow \qquad x = y (a+x²)
$$
E da equação (2) do item anterior, temos:
$$
b- ay - x²y = 0 \qquad \rightarrow \qquad y (a + x²) = b \qquad \rightarrow \qquad y = \frac{b}{a + x²}
$$


```python
#b)
def f(x,y):
    return y*(a + x**2)
def g(x):
    return b/(a + x**2)
    

a = 1
b = 2
x_0 = 2
y_0 = 3
tol = 1e-6

for i in range(100):
    x = f(x_0,y_0)
    y = g(x_0)
    print(x_0,y_0)
    x_0 = x
    y_0 = y

```
