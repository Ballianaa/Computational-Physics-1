```python
%matplotlib inline
from numpy import sin, sqrt
from gaussxw import gaussxw
from numpy import exp
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python

p_d = 6.022 * 10**28
theta = 428
V = 0.001
k_b = 1.38 * 10**-23 
T = 100

constante = 9 * V * p_d * k_b * (T/theta)**3

def f(x):
    return ((x**4) * exp(x)) / (exp(x) - 1)**2

a = theta/T
b = 0
N = 50
y,w = gaussxw(N)
x = ((b-a)/2)*y+(b+a)/2
wl = ((b-a)/2)*w
integ = 0.0

for k in range(N):
    integ+=wl[k]*f(x[k])

print("Para um número de fatias",N,"o valor estimado da integral é",integ * constante,".")
```

### Questão 3

```python
G = 6.674e-11  
sig = 100  

def f(x, y, z):
    return 1 / ((x**2 + y**2 + z**2)**1.5)

def F(z):
    N = 50
    a, b = -5, 5
    c = G * sig * z

    y, w = np.polynomial.legendre.leggauss(N)
    x = (((b - a)) * y + (b + a)) / 2
    wl = ((b - a) / 2) * w

    soma = 0
    for k in range(N):
        for i in range(N):
            soma += wl[k] * wl[i] * f(x[k], x[i], z)

    return c * soma

Z = np.linspace(0.3, 10, 100)

Fg = []
for z in Z:
    Fg.append(F(z))

print(Z)

plt.plot(Z, Fg, '.')
plt.xlabel("Z (m)")
plt.ylabel("Força gravitacional (N)")
plt.grid()
```
